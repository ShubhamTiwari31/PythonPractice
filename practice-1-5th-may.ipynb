```python
#increasing triangle
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#decreasing triangle
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#right increasing triangle
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#right decreasing triangle
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#hill pattern
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#decreasing hill pattern
n = int(input("Please enter a number for your pattern: "))
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n-1):
        print('*',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#deep and shallow copy
#shallow copy - change in one list will result to the change in other list too as they are pointing to the same location

lis1 = [1,2,3,4,5]
lis2 = [6,7,8,9,10]
lis1=lis2

lis1 #[6, 7, 8, 9, 10]
lis2 #[6, 7, 8, 9, 10]

lis1[0]=1 
lis1 #[1, 7, 8, 9, 10]
lis2 #[1, 7, 8, 9, 10] lis2 got changed even if we did not do anything with it, so this is shallow copy

```

```python
#deep copy
lis1 = [1,2,3,4,5]
lis2 = [6,7,8,9,10]

lis2 = lis1.copy() #copying lis1 values in lis2, all the lis1 values will be copied and lis2 will have same values but at a different memory location

lis2 #[1, 2, 3, 4, 5]

lis1 #[1, 2, 3, 4, 5]

lis1[0]=9
lis1 #[9, 2, 3, 4, 5]

lis2 #[1, 2, 3, 4, 5] lis2 will remain the same and there will be no change in lis2 even if we changed lis1 as we made a copy, so this is the deep copy

lis2[0]=10
lis2 #[10, 2, 3, 4, 5]

lis1 #[9, 2, 3, 4, 5] lis1 will remain same even if there is change in lis2 as they both are pointing to different memory locations and not same
```
