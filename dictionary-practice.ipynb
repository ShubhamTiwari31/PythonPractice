```python
#Dictionaries - is a type of data structure where the data is stored in the form of key value pair. Keys are unique and immutable, but dictionary as a whole is mutable.
```

```python
#Define
d={"Name":"Shubham","Email":"shubhamtiwari0131@gmail.com","Contact":901339}
d
```

```python
#access data - dictionary values can be accessed by its keys, the keys acts as indexes
d['Name']
```

```python
#Modifying values
d['Name']='Mann'
d
```

```python
'''Rules for defining keys

Keys can only be numerical or string and not any special characters.
Any mutable ds like list, set or dictionary itself can not be a key, as it has variable hash value. 
Tuple can be a key as it is immutable.
If there are two keys with the same name, the last or latest added key will overwrite the previous key.
'''
```

```python
#dictionary values can be a list or tuple
d={"names":['Shubham','Mann'],"Contact":(9999,8888)}
d
```

```python
#adding another key value pair
d['address']='Delhi'
d
```

```python
#deleting a key value pair
del d['address']
d
```

```python
#deleting via pop function - pop function returns the value which is popped too
d.pop('Contact')
```

```python
d
```

```python
#Merging a dictionary - via update function
d1={"name":"Shubh"}
d2={"Name":"Mann"}
d1.update(d2)
d1
```

```python
#zip function - attaches corresponding values with different keys
stu=['Shubh','Mann']
dob=[2002,2004]
dict={}

for i in zip(stu,dob):
    print(i)

#converting into dictionary
{i:j for i,j in zip(stu,dob)}

#revise again
```

```python
#example 2 of dictionary comprehension
user_id=[1,2,3]
user_name=['a21','a22','a23']
{i:j for i,j in zip(user_id,user_name)}
```

```python
#to see all key value pairs
d.items()
#iterate om the dictionary
for i,j in d.items():
    print(i,j)
#iterate on the keys only
for i in d.keys():
    print(i)
#iterate on the values only
for j in d.values():
    print(j)
```
