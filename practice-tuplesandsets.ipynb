Tuples - Tuples are the ordered collection of heterogeneous data and is immutable.

```python
#Defining a tuple
t = (1,1,2,3,4,5,2,4,5,'Hi','Hello')
t
```

```python
#Accessing tuple
t[0]
```

```python
#Modifying tuple - will throw error
t[0]=1
```

```python
#Tuple operations
#count a specific element

t.count(2)

#To find index of particular element
t.index(2) #returns the index of first occurence of the element

#slicing in tuple
t[0:9]

#reverse the tuple
t[::-1]

```

```python
#max and min value in tuple
max(t) #max value will not be given as there are strings and integer present, if there is only integer then it can give max value
min(t)
```

```python
t1=(1,2,3,4,5,6,1001,100000)
max(t1)
min(t1)
```

```python
#tuple inside tuple
t2=('a','b')
t3=(t1,t2)
t3
```

```python
#delete tuple 
t2
del t2
t2 #not present now
```

```python
#length of tuple
len(t1) #gives the total elements inside a tuple
```

**SETS** - Sets are the unordered collection of data. The unique thing about this data structure is that it doesn't allow duplicate values. Sets are mutable but if you want a data structure inside a set, that must be an immutable data structure i.e. tuple or strings and not list or set itself because immutable data structures have stable hash values and mutable data structures do not have stable hash values

```python
#Defining sets
s={1,2,3,'Hi','Hello'}
s
```

```python
#No accessing in sets as no indexing is there as it is unordered therefore no slicing too
s[0] #will throw an error
```

```python
#convert to a set - list can be converted in a set

l=[1,1,2,2,3,3,'Hi','Hi']
l = set(l)
type(l)
```

```python
#Data structures inside set - MUTABLE DS CAN NOT BE DEFINED INSIDE SETS AS THEY HAVE UNSTABLE HASH VALUES BUT IMMUTABLE DS CAN BE DEFINED INSIDE SET
s={1,2,3,4,(5,6,7),'Hi'} #tuples and strings can be defined inside set
s
```

```python
#Mutability of sets
S = {1,2,3,4}
S.add(100)
S
S.remove(100)
S
```

```python
#Update strings inside set
S
S.update("Shubham") #it is iteratable so the string will be updated one by one letter by letter, not necessary in same order
S
```

```python
#Update string as a whole - use list
S.update(['Shubham'])
S #using list to update will update our string as a whole
```

```python
#Updating integers - to update integers we have to use the list again
S.update([10,11])
S
```

```python
#clear set elements
S.clear()
S
```

```python
#Delete Set
del S
S #will throw not defined error as it is deleted now
```

```python
#Discard - deletes an element from set like remove, but if the element is not present in the set it doesn't throw an error unlike remove
s
s.discard('HELLO') #no error thrown even if 'HELLO' was not present in the set
```

```python
'''SET OPERATIONS - There are four set operations
1. Union | - Returns all value of both sets
2. Intersection & - Returns common value from both sets
3. Difference - - Returns all the value from first set (excludes common)
4. Symmetric Difference ^ - Returns the uncommon values from both sets '''

s1={1,2,3,4}
s2={3,4,5,6}
```

```python
#Union
s1|s2
```

```python
#intersection
s1&s2
```

```python
#difference
s1-s2
```

```python
#symmetric difference
s1^s2
```
