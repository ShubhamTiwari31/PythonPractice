```python
#patterns
#1 Square pattern
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(n):
        print('*',end=' ')
    print()
```

```python
#2. Increasing right angle triangle
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#3. Decreasing triangle pattern
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#4. Right Side Increasing Triangle
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#5. Left Downward Triangle
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#6. Hill Pattern
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#7. Reverse Hill
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n-1):
        print('*',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#8. Left Pascal
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n-1):
    for j in range(i+1):
        print('*',end=' ')
    print()
for i in range(n):
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#9. Right Pascal
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n-1):
    for j in range(i,n):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
for i in range(n):
    for j in range(i+1):
        print(' ',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#10. Butterfly
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n-1):
    for j in range(i+1):
        print('*',end=' ')
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
for i in range(n):
    for j in range(i,n):
        print('*',end=' ')
    for j in range(i):
        print(' ',end=' ')
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

Remake the butterfly, double hill, diamond, sandglass, etc. 

```python
#11. Double hill
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n):
    for j in range(i,n):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```

```python
#12. Diamond
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n-1):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
for i in range(n):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n-1):
        print('*',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
```

```python
#13. Sandglass
n=int(input("Enter a number for which you want to see the pattern: "))
for i in range(n-1):
    for j in range(i):
        print(' ',end=' ')
    for j in range(i,n-1):
        print('*',end=' ')
    for j in range(i,n):
        print('*',end=' ')
    print()
for i in range(n):
    for j in range(i,n-1):
        print(' ',end=' ')
    for j in range(i):
        print('*',end=' ')
    for j in range(i+1):
        print('*',end=' ')
    print()
```
